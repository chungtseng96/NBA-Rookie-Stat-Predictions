# Exploratory Data Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### College Trends

```python
file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//EDA datasets//full_college_stats.csv'
college = pd.read_csv(file)
college.head()
```

### Points

```python
plt.rcParams.update({'font.size': 16})
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs Points Scored')
sns.boxplot(college['year'], college['PTS'], ax=ax)
plt.xlabel('Year')
plt.ylabel('Points')
plt.show()
```

### Assist

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs Assists')
sns.boxplot(college['year'], college['AST'], ax=ax)
plt.xlabel('Year')
plt.ylabel('Assists')
plt.show()
```

### Rebounds 

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs Rebounds')
sns.boxplot(college['year'], college['TRB'], ax=ax)
plt.xlabel('Year')
plt.ylabel('Rebounds')
plt.show()
```
