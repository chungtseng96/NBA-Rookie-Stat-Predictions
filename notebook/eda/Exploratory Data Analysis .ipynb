# Exploratory Data Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Trend Analysis

## College Trends

```python
file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//EDA datasets//full_college_stats.csv'
college = pd.read_csv(file)
college.head()
```

```python
college.columns
```

### Points

```python
plt.rcParams.update({'font.size': 16})
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs Points Scored')
sns.boxplot(college['year'], college['PTS'], ax=ax)
sns.pointplot(x='year', y='PTS', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('Points')
plt.show()
```

### Assist

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs Assists')
sns.boxplot(college['year'], college['AST'], ax=ax)
sns.pointplot(x='year', y='AST', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('Assists')
plt.show()
```

### Rebounds 

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs Rebounds')
sns.boxplot(college['year'], college['TRB'], ax=ax)
sns.pointplot(x='year', y='TRB', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('Rebounds')
plt.show()
```

### 3-Pointers Attempted Per Games

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs 3PA')
sns.boxplot(college['year'], college['3PA'], ax=ax)
sns.pointplot(x='year', y='3PA', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('3PA')
plt.show()
```

### 3-Pointer Percentages 

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs 3P%')
sns.boxplot(college['year'], college['3P%'], ax=ax)
sns.pointplot(x='year', y='3P%', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.ylim(0.0,0.7)
plt.xlabel('Year')
plt.ylabel('3P%')
plt.show()
```

### Strength of Schedule

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs SOS')
sns.boxplot(college['year'], college['SOS'], ax=ax)
sns.pointplot(x='year', y='SOS', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('SOS')
plt.show()
```

### Turnover

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs TOV')
sns.boxplot(college['year'], college['TOV'], ax=ax)
sns.pointplot(x='year', y='TOV', data=college.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('TOV')
plt.show()
```
