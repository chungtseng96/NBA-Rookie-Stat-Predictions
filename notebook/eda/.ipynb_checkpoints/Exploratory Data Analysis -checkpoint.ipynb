# Exploratory Data Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//EDA datasets//full_college_stats.csv'
file = 'C:\\Users\\Chung-Tseng Wang\\Desktop\\NBA-Rookie-Stat-Predictions\\data\\processed\\final.csv'
data = pd.read_csv(file)
data.head()
```

```python
data.columns
```

### PER 

```python
plt.rcParams.update({'font.size': 16})
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Year vs PER')
sns.boxplot(data['year'], data['PER'], ax=ax)
sns.pointplot(x='year', y='PER', data=data.groupby('year', as_index=False).mean(), ax=ax)
plt.xlabel('Year')
plt.ylabel('Player Efficiency Rating')
plt.show()
```
