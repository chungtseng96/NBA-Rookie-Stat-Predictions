## Processing Scraped NBA Data

```python
import pandas as pd
import numpy as np
```

```python
#Import data and remove irrelevant columns
file = 'C:\\Users\\Chung-Tseng Wang\\Desktop\\Github Repo\\Predicting-NBA-Rookie-Stats\\data\\raw\\nba_stats_raw.csv'
raw = pd.read_csv(file)
raw = raw.drop(['Debut','MP','FG','FGA','3P','FT','FTA','ORB','TRB','AST','PF'],axis = 1)
raw.head()
```

```python
#Calculate steals, blocks, and turnover per game
raw['STL/G'] = raw['STL'] / raw['G']
raw['BLK/G'] = raw['BLK'] / raw['G']
raw['TOV/G'] = raw['TOV'] / raw['G']
raw = raw.drop(['Yrs','3PA','STL','BLK','TOV','PTS','MP.1','id'], axis = 1)
raw.head()
```

```python
raw.info()
```

```python
raw.describe()
```

```python
raw.isna().sum()
```

```python
raw['FG%'].fillna((raw['FG%'].mean()),inplace = True)
raw['3P%'].fillna((raw['FG%'].mean()),inplace = True)
raw['FT%'].fillna((raw['FG%'].mean()),inplace = True)
raw.head()
```

```python
raw.isna().sum()
```

```python
from unidecode import unidecode
raw['Player'] = raw['Player'].apply(unidecode)
raw.head()
```

```python
raw = raw.rename(columns = {'G':'NBA_G','FG%':'NBA_FGP','3P%':'NBA_3PP','FT%':'NBA_FTP','PS/G':'NBA_PTS','TRB/G':'NBA_TRB','AST/G':'NBA_AST','STL/G':'NBA_STL','BLK/G':'NBA_BLK','TOV/G':'NBA_TOV'})
raw.head()
```

```python
raw.to_csv('processed_nba.csv',index = True)
```

## Processing Scraped College Data

```python
raw_c = pd.read_csv('C:\\Users\\Chung-Tseng Wang\\Desktop\\Github Repo\\Predicting-NBA-Rookie-Stats\\data\\raw\\college_stats_raw.csv') 
raw_c.head()
```

```python
raw_c.describe()
```

```python
raw_c.info()
```

```python
raw_c.isna().sum()
```

```python
raw_c['FGP'].fillna((raw_c['FGP'].mean()),inplace = True)
raw_c['3P%'].fillna((raw_c['3P%'].mean()),inplace = True)
raw_c['FTP'].fillna((raw_c['FTP'].mean()),inplace = True)
raw_c['TOV'].fillna((raw_c['TOV'].mean()),inplace = True)
```

```python
raw_c.isna().sum()
```

```python
raw_c['Player'] = raw_c['Player'].apply(unidecode)
raw_c.head()
```

```python
raw_c = raw_c.rename(columns = {'G':'c_G','FGP':'c_FGP','3P%':'c_3PP','FTP':'c_FTP','PTS':'c_PTS','TRB':'c_TRB','AST':'c_AST','STL':'c_STL','BLK':'c_BLK','TOV':'c_TOV'})
raw_c.head()
```

```python
raw_c.to_csv('processed_college.csv',index = True)
```

```python
final = raw.merge(raw_c, how = 'left', left_on = raw['Player'], right_on = raw_c['Player'])
final.head()
```

```python
final.isna().sum()
```

```python
final.to_csv('final_training_set.csv',index = True)
```
