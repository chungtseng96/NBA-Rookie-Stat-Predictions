## Processing Scraped NBA Data

```python
import pandas as pd
import numpy as np
import os 
```

```python
#Import data and remove irrelevant columns
file = 'C:\\Users\\Chung-Tseng Wang\\Desktop\\Github Repo\\Predicting-NBA-Rookie-Stats\\data\\raw\\nba_stats_raw.csv'
#file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//raw//nba_stats_raw.csv'
nba = pd.read_csv(file)
nba = nba.drop(['Debut','MP','FG','FGA','3P','FT','FTA','ORB','TRB','AST','PF'],axis = 1)
nba.head()
```

```python
#Calculate steals, blocks, and turnover per game
nba['STL/G'] = nba['STL'] / nba['G']
nba['BLK/G'] = nba['BLK'] / nba['G']
nba['TOV/G'] = nba['TOV'] / nba['G']
nba = nba.drop(['Yrs','3PA','STL','BLK','TOV','PTS','MP.1','id'], axis = 1)
nba.head()
```

```python
nba.info()
```

```python
nba.describe()
```

```python
nba.isnull().sum()
```

```python
nba['FG%'].fillna((nba['FG%'].mean()),inplace = True)
nba['3P%'].fillna((nba['3P%'].mean()),inplace = True)
nba['FT%'].fillna((nba['FT%'].mean()),inplace = True)
nba.head()
```

```python
nba.isnull().sum()
```

```python
from unidecode import unidecode
nba['Player'] = nba['Player'].apply(unidecode)
nba.head()
```

```python
nba = nba.rename(columns = {'G':'NBA_G','FG%':'NBA_FGP','3P%':'NBA_3PP','FT%':'NBA_FTP','PS/G':'NBA_PTS','TRB/G':'NBA_TRB','AST/G':'NBA_AST','STL/G':'NBA_STL','BLK/G':'NBA_BLK','TOV/G':'NBA_TOV'})
nba.head()
```

```python
nba.to_csv('processed_nba.csv',index = True)
```

## Processing Scraped College Data

```python
file = 'C:\\Users\\Chung-Tseng Wang\\Desktop\\NBA-Rookie-Stat-Predictions\\data\\raw\\combined_college_stats.csv'
#file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//raw//college_stats_raw.csv'
college = pd.read_csv(file)
college.head()
```

```python
college.describe()
```

```python
college.info()
```

```python
college.isnull().sum()
```

```python
null_list = ['2PP', '3P%', 'DRB','FGP','GS','MP','ORB','SOS','TOV','eFGP','PER','WS']
for n in null_list:
    college[n] = college[n].fillna(college[n].mean())
#college.fillna(college.mean, inplace = True)
```

```python
college.isnull().sum()
```

```python
college['Player'] = college['Player'].apply(unidecode)
college.head()
```

```python
college.to_csv('combined_processed_college.csv',index = False)
```

```python
college_2020 = college[college['year'] == 2020]
college_2020.to_csv('college_2020_processed.csv', index = False)
college_past = college[college['year'] != 2020]
college_past.to_csv('college_past_processed.csv', index = False)
```
