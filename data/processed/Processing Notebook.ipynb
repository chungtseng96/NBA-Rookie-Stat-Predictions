## Processing Scraped NBA Data

```python
import pandas as pd
import numpy as np
```

```python
#Import data and remove irrelevant columns
file = 'C:\\Users\\Chung-Tseng Wang\\Desktop\\Github Repo\\Predicting-NBA-Rookie-Stats\\data\\raw\\nba_stats_raw.csv'
raw = pd.read_csv(file)
raw = raw.drop(['Debut','MP','FG','FGA','3P','FT','FTA','ORB','TRB','AST','PF'],axis = 1)
raw.head()
```

```python
#Calculate steals, blocks, and turnover per game
raw['STL/G'] = raw['STL'] / raw['G']
raw['BLK/G'] = raw['BLK'] / raw['G']
raw['TOV/G'] = raw['TOV'] / raw['G']
raw = raw.drop(['Yrs','3PA','STL','BLK','TOV','PTS','MP.1','id'], axis = 1)
raw.head()
```

```python
raw.info()
```

```python
raw.describe()
```

```python
raw.isna().sum()
```

```python
raw['FG%'].fillna((raw['FG%'].mean()),inplace = True)
raw['3P%'].fillna((raw['3P%'].mean()),inplace = True)
raw['FT%'].fillna((raw['FT%'].mean()),inplace = True)
raw.head()
```

```python
raw.isna().sum()
```

```python
from unidecode import unidecode
raw['Player'] = raw['Player'].apply(unidecode)
raw.head()
```

```python
raw = raw.rename(columns = {'G':'NBA_G','FG%':'NBA_FGP','3P%':'NBA_3PP','FT%':'NBA_FTP','PS/G':'NBA_PTS','TRB/G':'NBA_TRB','AST/G':'NBA_AST','STL/G':'NBA_STL','BLK/G':'NBA_BLK','TOV/G':'NBA_TOV'})
raw.head()
```

```python
raw.to_csv('processed_nba.csv',index = True)
```
