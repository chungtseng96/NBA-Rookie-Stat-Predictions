Scraping College Stats

```python
import pandas as pd
import numpy as np
import requests
import lxml.html as lh
from bs4 import BeautifulSoup
import re 
```

```python
file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//processed//processed_nba.csv'
nba = pd.read_csv(file)
nba.head()
```

```python
players_list = nba['Player'].values
len(players_list)
```

```python
data = []
for player in players_list:
    url_name = player
    player = player.lower()
    player = player.replace('.','')
    player = player.replace("'",'')
    player = player.replace(' ','-')
    website_url = requests.get('https://www.sports-reference.com/cbb/players/'+ player +'-1.html')
    soup = BeautifulSoup(website_url.content,'lxml')
    table = soup.find('table',{'id':'players_per_game'})
    if table is not None:
        data = data + [url_name]
        box = table.findAll('td')
        box_list = []
        for n in box:
            box_list.append(n.get_text())
        data = data + box_list[-28:]
```

```python
transformed = np.reshape(np.array(data),(-1,29))
```

```python
#To get column headers
headers = ['Player','School','Conf','G','GS','MP','FG','FGA','FGP','2P','2PA','2PP','3P','3PA','3P%','FT','FTA','FTP','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','nan','SOS']
college = pd.DataFrame(transformed, columns = headers)
college = college.drop(['Conf','nan','PF'], axis = 1)
college.head()
```

### Scraping advance stats

```python
players_list = full_college_stats['Player'].values
```

```python
data = []
for player in players_list:
    url_name = player
    player = player.lower()
    player = player.replace('.','')
    player = player.replace("'",'')
    player = player.replace(' ','-')
    website_url = requests.get('https://www.sports-reference.com/cbb/players/'+ player +'-1.html')
    soup = BeautifulSoup(website_url.content,'lxml')
    #Position 
    div = soup.find_all('div',{'class' : 'nothumb'})
    for x in div:
        ad = x.find_all('p')
        position = (ad[0].text.replace('Position:', "").strip())
    #eFG%
    div2 = soup.find_all('div',{'class' : 'p2'})
    for x in div2:
        ad = x.find_all('p')
        eFG = ad[-1].text
    #PER and WS
    div3 = soup.find_all('div',{'class' : 'p3'})
    for x in div3:
        ad = x.find_all('div')
        if len(ad) >= 1: 
            aPER = ad[0].find_all('p')
            aWS = ad[-1].find_all('p')
            PER = []
            for n in aPER:
                PER.append(n.text)
            WS = []
            for n in aWS:
                WS.append(n.text)
            cPER = PER[-1]
            cWS = WS[-1]
        else:
            cPER = 0
            cWS = 0
    data = data + [url_name] + [position] + [eFG] + [cPER] + [cWS]
```

```python
transformed2 = np.reshape(np.array(data),(-1,5))
```

```python
headers = ['Player','Position','eFGP','PER','WS']
advance = pd.DataFrame(transformed2, columns = headers)
advance.head()
```

```python
full_college_stats = full_college_stats.merge(advance, on = 'Player', how = 'left')
full_college_stats[full_college_stats['year'] == 2020].head()
```

```python
college_stats_2020 = full_college_stats[full_college_stats['year'] == 2020]
past_college_stats = full_college_stats[full_college_stats['year'] != 2020]
full_college_stats.to_csv('combined_college_stats.csv')
college_stats_2020.to_csv('2020_college_stats.csv')
past_college_stats.to_csv('past_college_stats.csv')
```
