Scraping International Player Stats

```python
import pandas as pd
import numpy as np
import requests
import lxml.html as lh
from bs4 import BeautifulSoup
```

```python
#file = 'C:\\Users\\Chung-Tseng Wang\\Desktop\\Github Repo\\Predicting-NBA-Rookie-Stats\\data\\processed\\final_training_set.csv'
file = '//Users//chung-tsengwang//Desktop//Github Repo//Predicting-NBA-Rookie-Stats//data//processed//final_training_set.csv'
nba = pd.read_csv(file, index_col = 'Player')
nba.head()
```

```python
players_list = nba[nba['School'].isnull()].index.values
len(players_list)
```

```python
players_list
```

```python
data = []
for player in players_list:
    url_name = player
    player = player.lower()
    player = player.replace('.','')
    player = player.replace("'",'')
    player = player.replace(' ','-')
    website_url = requests.get('https://www.basketball-reference.com/international/players/'+ player +'-1.html')
    soup = BeautifulSoup(website_url.content,'lxml')
    table = soup.find('table',{'id':'per_game-clu-reg'})
    if table is not None:
        data = data + [url_name]
        stats = table.findAll('td')
        elements = []
        for n in stats:
            elements.append(n.get_text())
        data = data + elements[-22:]
```

```python
transformed = np.reshape(np.array(data),(-1,23))
```

```python
#To get column headers
headers = ['Player','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS']
inter = pd.DataFrame(transformed, columns = headers)
inter = inter.drop(['MP','FG','FGA','3P','3PA','2P','2PA','2P%','FT','FTA','ORB','DRB','PF'], axis = 1)
inter.index = inter['Player']
inter = inter.drop(['Player'],axis = 1)
inter['3P%'] = inter['3P%'].replace('',0)
inter['FT%'] = inter['FT%'].replace('',0)
inter.head()
```

```python
inter.info()
```

```python
inter.describe()
```

```python
for n in list(inter.columns):
    inter[n] = inter[n].astype(float)
```

```python
inter.info()
```

```python
inter = inter.rename(columns = {'G':'c_G','FG%':'c_FGP','3P%':'c_3PP','FT%':'c_FTP','PTS':'c_PTS','TRB':'c_TRB','AST':'c_AST','STL':'c_STL','BLK':'c_BLK','TOV':'c_TOV'})
inter.head()
```

```python
inter.to_csv('processed_inter.csv', index = True)
```
